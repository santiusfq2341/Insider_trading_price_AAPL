```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# lectura de datos de tablas fct
fct_precios_insiders_AAPL = pd.read_csv(('../Dataset/clean/AAPL_fct_precios_insider.csv'))
fct_precios_insiders_AAPL.head()
```

```python
# eliminar el simbolo % y el guion - de la columna "delta_owned".
fct_precios_insiders_AAPL['delta_owned'] = fct_precios_insiders_AAPL['delta_owned'].str.replace('%', '').str.replace('-', '')
```

```python
fct_precios_insiders_AAPL
```

# COMENTARIO
Necesitaba quitar el símbolo de % de la columna delta_owned porque quiero realizar una métrica para saber si el insider vendió una parte significativa de sus acciones. Por ejemplo, lo quiero estimar en un umbral del 50%.

Hay que entender que la columna delta_owned representa el % de acciones que tiene esa persona luego de haber realizado una transacción, por ejemplo, si es venta el delta_owned será una disminución, si es compra será un aumento. 

Para el caso de las personas que su delta_owned es 0, no quiere decir que no hicieron transacciones, quiere decir que el porcentaje negociado es ínfimo frente al total de las acciones que poseen. Esto se puede evidenciar en la línea 220 de la tabla fct.

```python
# transformar la columna "delta_owned" a tipo int 
fct_precios_insiders_AAPL['delta_owned'] = fct_precios_insiders_AAPL['delta_owned'].astype(int)
```

```python
fct_precios_insiders_AAPL
```

```python
# crear una nueva columna para determinar si el insider ha vendido o comprado en base al valor de la columna "delta_owned" mayor al 10% debe colocar 1 es venta significativa y 0 si no lo es

fct_precios_insiders_AAPL['significant_transaction'] = np.where(fct_precios_insiders_AAPL['delta_owned'] >= 10, 1, 0)
fct_precios_insiders_AAPL.head(10)
```

```python
# quiero filtrar los datos que muestre los 1 en significant_transaction y los valores menores a 0 en movimiento

filtered_data = fct_precios_insiders_AAPL[
	(fct_precios_insiders_AAPL['significant_transaction'] == 1) & 
	(fct_precios_insiders_AAPL['movimiento'] < 0)
]
filtered_data
```

```python
filtered_data_2 = fct_precios_insiders_AAPL[
	(fct_precios_insiders_AAPL['significant_transaction'] == 1) & 
	(fct_precios_insiders_AAPL['movimiento'] > 0)
]
filtered_data_2
```

```python
# crear columna para determinar el impacto negativo si el insider ha vendido en base al valor de movimiento y significant_transaction, que el output sea true o false, con inplace=true
fct_precios_insiders_AAPL['impacto_negativo'] = fct_precios_insiders_AAPL.apply(
    lambda x: True if (x['movimiento'] < 0 and x['significant_transaction'] == 1) else False, axis=1
)
```

```python
fct_precios_insiders_AAPL
```

```python
fct_precios_insiders_AAPL['año'] = pd.to_datetime(fct_precios_insiders_AAPL['trade_date']).dt.year
fct_precios_insiders_AAPL['mes'] = pd.to_datetime(fct_precios_insiders_AAPL['trade_date']).dt.month
```

```python
fct_precios_insiders_AAPL
```

# COMENTARIO
Determiné un umbral del 10% porque tomando en cuenta que son directores de la compañía tienen mucha influencia y un movimiento de acciones por más pequeño que sea debe ser considerado como una alerta. 
Mi análisis será a partir de verificar el impacto con el cambio del precio de la acción en negativo, es decir, si un director que haya vendido por lo menos el 10% de sus acciones afecta al precio de la acción en negativo, se consideraría un resultado positivo.
